```python
## Scraping to get all links
import bs4 as bs
from urllib.request import urlopen
#import lxml.html
import pandas as pd
import os
import requests
import numpy as np
import re
import nltk
from nltk import word_tokenize
from nltk.tokenize import sent_tokenize
```

```python
url = "http://www.mtsamples.com/site/pages/browse.asp?type=97-Consult%20-%20History%20and%20Phy."
x = urlopen(url)
x = urlopen("http://www.mtsamples.com/site/pages/browse.asp?type=97-Consult%20-%20History%20and%20Phy.")
new = x.read()
links = []
new
```

```python
f = requests.get(url)
print (f.text)
 
new = f.text
soup = bs.BeautifulSoup(new,"lxml")
```

```python
for link in soup.find_all('a',href = True):
    links.append((link.get('href')))
 
print(links)
```

```python
#### Sort all links so that we can get the 516 patient record links
links.sort()
print(links)
 
df1 = pd.DataFrame({'links': links})
```
